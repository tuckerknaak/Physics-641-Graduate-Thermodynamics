### Problem 4.14.  Density of states of an ideal gas.

#### Calculate the density of states $g(E,V,N)$ and verifty that $\Gamma(E,V,N)$ and $g(E,V,N)$ are rapidly increasing functions of $E$, $V$, $N$.

```python
'''Required Libraries'''
import matplotlib.pyplot as plt
import numpy as np
```

#### Note that $\Gamma(E,V,N)=A\left(\frac{V}{N}\right)^N\left(\frac{E}{N}\right)^{3N/2}\text{e}^{5N/2}\text{ and }g(E,V,N)=\frac{d\Gamma}{dE}=\frac{3A}{2}\left(\frac{V}{N}\right)^N\left(\frac{E}{N}\right)^{3N/2-1}\text{e}^{5N/2}$.

```python
'''Function to calculate gamma'''
def gamma(A: float, E: float, V: float, N: float):
    T = A * ((V / N)**N) * ((E / N)**(3 * N / 2)) * np.exp(5 * N / 2)
    return T

'''Function to calculate the density of states'''
def density(A: float, E: float, V: float, N: float):
    g = (3 * A / 2) * ((V / N)**N) * ((E / N)**((3 * N / 2) - 1)) * np.exp(5 * N / 2)
    return g
```

#### (a) Gamma and Density of States for increasing E.

```python
'''Set range of energies to consider'''
energy = np.linspace(0.01, 10, 1000, endpoint = True)

'''Calculate Gamma for different limits'''
gamma1_list = [gamma(1, E, 1, 1) for E in energy]       #V = N = 1
gamma2_list = [gamma(1, E, 1, 20) for E in energy]      #V = 1, N >> E
density1_list = [density(1, E, 1, 1) for E in energy]   #V = N = 1
density2_list = [density(1, E, 1, 20) for E in energy]  #V = 1, N >> E

'''Plot results'''
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (16, 10), gridspec_kw = {'width_ratios': [1, 1],
                                                                                      'height_ratios': [1, 1]})
fig.tight_layout(pad = 5)
for ax in [ax1, ax2, ax3, ax4]:
    ax.set_xlabel('Volume $V$')
    ax.grid(True, linestyle = 'dashed', color = 'darkgray', alpha = 0.25)
    
'''Plot 1: Gamma V = N = 1'''
ax1.set_ylabel('$\Gamma(E,V,N)$')
ax1.set_title('Gamma for $V=N=1$')
ax1.plot(energy, gamma1_list, linestyle = 'solid', color = 'red')

'''Plot 2: Density V = N = 1'''
ax2.set_ylabel('$g(E,V,N)$')
ax2.set_title('Density of States for $V=N=1$')
ax2.plot(energy, density1_list, linestyle = 'solid', color = 'mediumorchid')

'''Plot 3: Gamma V = 1, N >> E'''
ax3.set_ylabel('$\Gamma(E,V,N)$')
ax3.set_title('Gamma for $N>>E,\ V=1$')
ax3.plot(energy, gamma2_list, linestyle = 'solid', color = 'dodgerblue', label = '$N=20$')
ax3.legend(loc = 'upper left', prop = {'size': 15})

'''Plot 4: Density V = 1, N >> E'''
ax4.set_ylabel('$g(E,V,N)$')
ax4.set_title('Density of States for $N>>E,\ V=1$')
ax4.plot(energy, density2_list, linestyle = 'solid', color = 'limegreen', label = '$N=20$')
ax4.legend(loc = 'upper left', prop = {'size': 15})

'''Save figure'''
fig.savefig('c:/Users/Tucker Knaak/Downloads/Gamma_Density_E.png')
```
